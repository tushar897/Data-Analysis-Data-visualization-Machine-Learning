```python

import pandas as pd


df_items = pd.read_csv('E:\\cleaning_data\\shops\\items.csv')
df_items_cat = pd.read_csv('E:\\cleaning_data\\shops\\item_categories.csv')
df_train = pd.read_csv('E:\\cleaning_data\\shops\\sales_train.csv')
df_subs = pd.read_csv('E:\\cleaning_data\\shops\\sample_submission.csv')
df_shops = pd.read_csv('E:\\cleaning_data\\shops\\shops.csv')

df_items.head()
```

```python
train = df_train.join(df_shops,on='shop_id',rsuffix='_').join(df_items,on='item_id',rsuffix='_').join(df_items_cat,on='item_category_id',rsuffix='_').drop(['item_category_id','item_id','shop_id'],axis=1)
train.head()
```
