```python
import pandas as pd

df_vaccine = pd.read_csv('E:\\cleaning_data\\corona\\covid_vaccine_statewise.csv')
df_covid = pd.read_csv('E:\\cleaning_data\\corona\\covid_19_india.csv')
df_testing = pd.read_csv('E:\\cleaning_data\\corona\\StatewiseTestingDetails.csv')


df_covid.head()
```

```python
df_duplicates = df_covid[df_covid.duplicated()]
```

```python
df_deathsovertime = df_covid.groupby(['State/UnionTerritory'])['Deaths','Cured'].sum().reset_index().sort_values('State/UnionTerritory',ascending=True)

import seaborn as sns
import matplotlib.pyplot as plt

sns.set(rc={'figure.figsize':(20,20)})
plt.xticks(rotation=45)

sns.barplot(x='State/UnionTerritory',y='Cured',data = df_deathsovertime)
```

```python
import numpy as np
df_sampleoverstate = df_testing.groupby(['State'])['TotalSamples'].sum().reset_index().sort_values('State',ascending=True)

sns.set(rc={'figure.figsize':(20,20)})
plt.xticks(rotation=45)

X_axis = np.arange(len(df_sampleoverstate['State']))
x_axis = np.arange(len(df_deathsovertime['State/UnionTerritory']))

sns.barplot(y='TotalSamples',data = df_sampleoverstate,label='samples')
sns.barplot(y='Cured',data = df_deathsovertime,label='cured')


```
